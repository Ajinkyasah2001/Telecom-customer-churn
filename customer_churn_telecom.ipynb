## Feature Importance & Reduction Analysis

### Methods to Reduce Features:
1. **Statistical correlation** - drop highly correlated features
2. **Tree-based importance** - RandomForest/XGBoost feature importance
3. **Mutual information** - measure feature-target relationship
4. **VIF (Variance Inflation Factor)** - detect multicollinearity
5. **Domain knowledge** - drop redundant/leakage features


```python
# 1. FEATURE IMPORTANCE using RandomForest
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python

# Prepare data
X = df_cleaned.drop('Churn', axis=1)
y = df_cleaned['Churn']

# Encode categorical columns
X_copy = X.copy()
for col in X_copy.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X_copy[col] = le.fit_transform(X_copy[col])

# Train RandomForest
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, max_depth=12)
rf.fit(X_copy, y)

# Get feature importance
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
}).sort_values('Importance', ascending=False)

# Plot top 20 features
plt.figure(figsize=(10, 8))
sns.barplot(data=importance_df.head(20), x='Importance', y='Feature', palette='viridis')
plt.title('Top 20 Most Important Features (RandomForest)', fontsize=14, fontweight='bold')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()

print("=" * 60)
print("TOP 15 IMPORTANT FEATURES")
print("=" * 60)
print(importance_df.head(15).to_string(index=False))
```

```python
imp_feature_rf = importance_df.head(30)
imp_feature_rf
```

```python
# 2. MULTICOLLINEARITY CHECK - VIF (Variance Inflation Factor)
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Use only numeric columns
X_numeric = X_copy.select_dtypes(include=[np.number])

# Calculate VIF for each feature
vif_data = pd.DataFrame({
    'Feature': X_numeric.columns,
    'VIF': [variance_inflation_factor(X_numeric.values, i) for i in range(X_numeric.shape[1])]
}).sort_values('VIF', ascending=False)

print("=" * 60)
print("VARIANCE INFLATION FACTOR (VIF)")
print("VIF > 10 indicates high multicollinearity (consider dropping)")
print("=" * 60)
print(vif_data.head(20).to_string(index=False))

# Features with high VIF
high_vif = vif_data[vif_data['VIF'] > 10]['Feature'].tolist()
print(f"\n⚠️  High VIF features (>10): {high_vif}")
```

```python
# 3. CORRELATION ANALYSIS - Find Redundant Features
corr_matrix = X_numeric.corr()


# Find feature pairs with high correlation (>0.9)
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > 0.9:
            high_corr_pairs.append({
                'Feature1': corr_matrix.columns[i],
                'Feature2': corr_matrix.columns[j],
                'Correlation': corr_matrix.iloc[i, j]
            })

high_corr_df = pd.DataFrame(high_corr_pairs).sort_values('Correlation', key=abs, ascending=False)

print("=" * 60)
print("HIGHLY CORRELATED FEATURE PAIRS (|correlation| > 0.9)")
print("Consider dropping one feature from each pair")
print("=" * 60)
if len(high_corr_df) > 0:
    print(high_corr_df.to_string(index=False))
else:
    print("No feature pairs with correlation > 0.9 found")

# Plot correlation heatmap for top 15 important features
top_features = importance_df.head(15)['Feature'].tolist()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix.loc[top_features, top_features], annot=True, fmt='.2f', 
            cmap='coolwarm', center=0, square=True, cbar_kws={'label': 'Correlation'})
plt.title('Correlation Matrix - Top 15 Important Features', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 4. MUTUAL INFORMATION - Feature-Target Dependency

from sklearn.feature_selection import mutual_info_classif

# Calculate mutual information
mi_scores = mutual_info_classif(X_copy, y, random_state=42)
mi_df = pd.DataFrame({
    'Feature': X.columns,
    'Mutual_Information': mi_scores
}).sort_values('Mutual_Information', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(data=mi_df.head(20), x='Mutual_Information', y='Feature', palette='plasma')
plt.title('Top 20 Features by Mutual Information with Target', fontsize=14, fontweight='bold')
plt.xlabel('Mutual Information Score')
plt.tight_layout()
plt.show()

print("=" * 60)
print("TOP 15 FEATURES BY MUTUAL INFORMATION")
print("=" * 60)
print(mi_df.head(15).to_string(index=False))
```

```python
def preprocess_test_data(test_df, features_to_keep=None, columns_to_drop=None, target_col='Churn'):
    """
    Preprocess test data to match the training data preprocessing pipeline.
    
    Parameters:
    -----------
    test_df : pd.DataFrame
        Raw test dataframe to be preprocessed
    features_to_keep : list, optional
        List of feature names to keep (excluding target). If None, applies standard dropping.
    columns_to_drop : list, optional
        List of columns to drop. If None, uses default dropped columns.
    target_col : str
        Name of the target column (default: 'Churn')
    
    Returns:
    --------
    pd.DataFrame
        Preprocessed test dataframe ready for model prediction
    """
    
    # Create a copy to avoid modifying original
    df_test = test_df.copy()
    
    # Step 1: Drop columns that were dropped from training data
    if columns_to_drop is None:
        columns_to_drop = ['CustomerID', 'AgeHH2', 'HandsetPrice', 'NotNewCellphoneUser', 
                          'ServiceArea', 'MaritalStatus', 'Homeownership']
    
    # Only drop columns that exist in the test dataframe
    cols_to_drop_exist = [col for col in columns_to_drop if col in df_test.columns]
    df_test.drop(columns=cols_to_drop_exist, inplace=True)
    
    print(f"✓ Dropped {len(cols_to_drop_exist)} columns: {cols_to_drop_exist}")
    print(f"  Shape after dropping columns: {df_test.shape}")
    
    # Step 2: Drop NaN rows
    rows_before = len(df_test)
    df_test.dropna(inplace=True)
    rows_after = len(df_test)
    print(f"\n✓ Dropped {rows_before - rows_after} rows with NaN values")
    print(f"  Shape after dropping NaN: {df_test.shape}")
    
    # Step 3: Drop rows with negative values in non-negative columns
    non_negative_cols = [
        "MonthlyRevenue", "MonthlyMinutes", "TotalRecurringCharge", "DirectorAssistedCalls",
        "OverageMinutes", "RoamingCalls", "DroppedCalls", "BlockedCalls", "UnansweredCalls",
        "CustomerCareCalls", "ThreewayCalls", "ReceivedCalls", "OutboundCalls", "InboundCalls",
        "PeakCallsInOut", "OffPeakCallsInOut", "DroppedBlockedCalls", "CallForwardingCalls",
        "CallWaitingCalls", "MonthsInService", "UniqueSubs", "ActiveSubs", "Handsets",
        "HandsetModels", "CurrentEquipmentDays", "AgeHH1", "RetentionCalls",
        "RetentionOffersAccepted", "ReferralsMadeBySubscriber", "AdjustmentsToCreditRating"
    ]
    
    non_negative_cols = [c for c in non_negative_cols if c in df_test.columns]
    neg_mask = (df_test[non_negative_cols] < 0).any(axis=1)
    rows_removed = neg_mask.sum()
    df_test = df_test.loc[~neg_mask].copy()
    df_test.reset_index(drop=True, inplace=True)
    
    if rows_removed > 0:
        print(f"\n✓ Dropped {rows_removed} rows with invalid negative values")
        print(f"  Shape after removing negatives: {df_test.shape}")
    
    # Step 4: Encode boolean categorical columns (Yes/No) to 1/0
    def convert_to_bool(df, column_name):
        df[column_name] = df[column_name].map({'Yes': 1, 'No': 0})
    
    # Identify and encode binary categorical columns
    bool_cols = []
    num_cols = df_test.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = [c for c in df_test.columns if c not in num_cols and c != target_col]
    
    for col in cat_cols:
        if df_test[col].nunique() == 2:
            convert_to_bool(df_test, col)
            bool_cols.append(col)
    
    if bool_cols:
        print(f"\n✓ Converted {len(bool_cols)} binary categorical columns to 0/1")
        print(f"  Columns: {bool_cols}")
    
    # Step 5: Label encode multi-class categorical columns
    from sklearn.preprocessing import LabelEncoder
    
    multi_cat_cols = [c for c in cat_cols if c not in bool_cols and c != target_col]
    
    le_dict = {}
    for col in multi_cat_cols:
        if col in df_test.columns:
            le = LabelEncoder()
            df_test[col] = le.fit_transform(df_test[col].astype(str))
            le_dict[col] = le
    
    if multi_cat_cols:
        print(f"\n✓ Label encoded {len(multi_cat_cols)} multi-class categorical columns")
        print(f"  Columns: {multi_cat_cols}")
    
    # Step 6: Keep only the specified features (if provided)
    if features_to_keep is not None:
        features_available = [f for f in features_to_keep if f in df_test.columns]
        df_test = df_test[features_available]
        print(f"\n✓ Selected {len(features_available)} specified features")
        print(f"  Final shape: {df_test.shape}")
    else:
        print(f"\n✓ Final shape: {df_test.shape}")
    
    print("\n" + "="*60)
    print("✅ TEST DATA PREPROCESSING COMPLETE")
    print("="*60)
    
    return df_test
```
